```python
import pandas as pd
```

```python
# Step 1: Data Ingestion - Load the input CSV file
def load_data(file_path):
    try:
        data = pd.read_csv(file_path)
        print("Data loaded successfully!")
        return data
    except FileNotFoundError:
        print(f"File not found: {file_path}")
        return None
    except Exception as e:
        print(f"Error loading data: {e}")
        return None

# Load the data
input_file = "C:/Users/ELCOT/Desktop/intern/input_data.csv"  # Path to the uploaded file
data = load_data(input_file)

# Display the first few rows of the data
if data is not None:
    display(data.head())
```

```python
# Step 2: Data Cleaning - Handle missing values and duplicates
def clean_data(data):
    if data is not None:
        # Drop rows with missing values
        data = data.dropna()
        # Drop duplicate rows
        data = data.drop_duplicates()
        print("Data cleaned successfully!")
        return data
    else:
        print("No data to clean!")
        return None

# Clean the data
data = clean_data(data)

# Display the cleaned data
if data is not None:
    display(data.head())
```

```python
# Step 3: Data Transformation - Add a new column (e.g., compute a metric)
def transform_data(data):
    if data is not None:
        # Example transformation: Add a new column called "Total" (sum of two numeric columns)
        if 'Column1' in data.columns and 'Column2' in data.columns:
            data['Total'] = data['Column1'] + data['Column2']
            print("Data transformed successfully!")
        else:
            print("Required columns not found for transformation.")
        return data
    else:
        print("No data to transform!")
        return None

# Transform the data
data = transform_data(data)

# Display the transformed data
if data is not None:
    display(data.head())
```

```python
# Step 4: Data Output - Save the processed data to a new CSV file
def save_data(data, output_file_path):
    if data is not None:
        try:
            data.to_csv(output_file_path, index=False)
            print(f"Data saved successfully to {output_file_path}!")
        except Exception as e:
            print(f"Error saving data: {e}")
    else:
        print("No data to save!")

# Save the data
output_file = "C:/Users/ELCOT/Desktop/intern/output_data.csv"  # Path to save the processed data
save_data(data, output_file)
```
